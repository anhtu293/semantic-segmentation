```python
catIDs = list(range(1,91))
y = np.zeros((image['width'], image['height'], 91))

for cat in catIDs:
    annIds = coco.getAnnIds(imgIds=image['id'],catIds=[cat])
    anns = coco.loadAnns(annIds)
    if len(annIds) > 0:
        for idx in annIds:       
            mask = coco.annToMask(coco.loadAnns([idx]))
            y[:,:,cat] = np.logical_and(y[:,:,cat], mask).astype(int)
```
